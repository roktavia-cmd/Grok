```python
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv('grok_ready.csv')
```

```python
semua_kata = " ". join(df['text_clean'].astype(str)).split()
hitung_kata = Counter(semua_kata)
```

```python
df_wordcloud = pd.DataFrame(hitung_kata.most_common(50), columns=['kata','Frekuensi'])
```

```python
df_wordcloud.to_csv('grok_wordcloud.csv', index=False)
print('OK')
```
