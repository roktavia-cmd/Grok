```python
import pandas as pd
from collections import Counter
```

```python
stopwords = set(['is', 'on', 'the', 'dont', 'to', 'a', 'an', 'and', 'in', 'if', 'for', 'you', 'your', 'with', 'it', 'this', 'that', 'of', 'be', 'are', 'was', 'were', 'my', 'me', 'at' ])
```

```python
df = pd.read_csv('grok_ready.csv')
```

```python
semua_kata = " ". join(df['text_clean'].astype(str)).lower().split()
```

```python
kata_bersih = [kata for kata in semua_kata if kata not in stopwords and len(kata) > 2]
hitung_kata = Counter(kata_bersih)
```

```python
df_wordcloud = pd.DataFrame(hitung_kata.most_common(50), columns=['kata','Frekuensi'])
```

```python
df_wordcloud.to_csv('grok_wordcloud.csv', index=False)
print('OK')
```
